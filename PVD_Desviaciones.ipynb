```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Leer los datos de Excel
# Asegúrate de reemplazar 'ruta_a_tu_archivo.xlsx' con la ruta real de tu archivo Excel
df = pd.read_excel('Muestreos.xlsx')

# Supongamos que los tamaños de las nanopartículas están en una columna llamada 'Tamaño'
tamanos = df['Length']

# Calcular la desviación estándar
desviacion_estandar = np.std(tamanos)

# Imprimir la desviación estándar
print(f'Desviación estándar: {desviacion_estandar}')

# Crear el histograma
plt.hist(tamanos, bins='auto', edgecolor='black')

# Añadir título y etiquetas
plt.title('Histograma de tamaños de nanopartículas')
plt.xlabel('Tamaño de nanopartículas')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()
```
